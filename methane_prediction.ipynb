```python
data['Date_Time'] = pd.to_datetime(data[['year', 'month', 'day', 'hour', 'minute', 'second']])
data = data.drop(columns = ['year', 'month', 'day', 'hour', 'minute', 'second'])
data
```

```python
dataset = data.copy()
dataset = dataset[(dataset['Date_Time'] >= '2014-3-29 14:44:57') & (dataset['Date_Time'] <= '2014-4-8 13:15:56')]
dataset
```

```python
dataset = dataset.drop(dataset[(dataset['MM256'] >= 2)].index)
dataset = dataset.drop(dataset[(dataset['MM256'] == 0.)].index)

dataset = dataset.drop(dataset[(dataset['AN311'] == 0.)].index)
dataset = dataset.drop(dataset[(dataset['TP1721'] == 0.)].index)
dataset = dataset.drop(dataset[(dataset['AN422'] < 0.5)].index)
```

```python
dataset['MM256_Filtered'] = dataset['MM256'].ewm(alpha=0.1).mean()
dataset['MM256_Filtered'] = dataset['MM256_Filtered'].ewm(alpha=0.01).mean()
```

```python
dataset = dataset.assign(number=range(1, len(dataset) + 1))
```

```python
scaler_1 = StandardScaler()

X_train_1_sec[columns] = scaler_1.fit_transform(X_train_1_sec[columns])
X_validation_1_sec[columns] = scaler_1.transform(X_validation_1_sec[columns])
X_test_1_sec[columns] = scaler_1.transform(X_test_1_sec[columns])

scaler_60 = StandardScaler()

X_train_60_sec[columns] = scaler_60.fit_transform(X_train_60_sec[columns])
X_validation_60_sec[columns] = scaler_60.transform(X_validation_60_sec[columns])
X_test_60_sec[columns] = scaler_60.transform(X_test_60_sec[columns])

scaler_900 = StandardScaler()

X_train_900_sec[columns] = scaler_900.fit_transform(X_train_900_sec[columns])
X_validation_900_sec[columns] = scaler_900.transform(X_validation_900_sec[columns])
X_test_900_sec[columns] = scaler_900.transform(X_test_900_sec[columns])
```

```python
import torch
import numpy as np

from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
class methaneDataset(Dataset):
    def __init__(self, x, y):
        self.x = torch.tensor(np.array(x), dtype=torch.float32)
        self.y = torch.tensor(np.array(y), dtype=torch.float32)

    def __len__(self):
        return self.x.shape[0]

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.act = nn.Sequential(nn.Linear(28, 57), nn.Tanh(),
                                 nn.Linear(57, 3), nn.Tanh(),
                                 nn.Linear(3, 1))

    def forward(self, x):
        return self.act(x)

def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    running_loss = 0.
    for batch, (X_for_train, y_for_train) in enumerate(dataloader):

        pred = model(X_for_train)

        loss = loss_fn(pred.squeeze(1), y_for_train)
        running_loss += loss.item()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 500 == 0:
            loss, current = loss.item(), batch * len(X_for_train)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
    train_losses.append(running_loss / len(dataloader))
    print(f"Train Error: \n Avg loss: {running_loss / len(dataloader):>8f} \n")


def validation(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0
    with torch.no_grad():
        for X_for_val, y_for_val in dataloader:
            pred = model(X_for_val)

            test_loss += loss_fn(pred.squeeze(1), y_for_val).item()

    test_loss /= num_batches
    test_losses.append(test_loss)
    print(f"Validation Error: \n Avg loss: {test_loss:>8f} \n")

def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0
    with torch.no_grad():
        for X_for_test, y_for_test in dataloader:
            pred = model(X_for_test)

            test_loss += loss_fn(pred.squeeze(1), y_for_test).item()

    test_loss /= num_batches
    print(f"Test Error: \n Avg loss: {test_loss:>8f} \n")
    return test_loss

dataset_train = methaneDataset(X_train_1_sec, y_train_1_sec)
dataset_validation = methaneDataset(X_validation_1_sec, y_validation_1_sec)
dataset_test = methaneDataset(X_test_1_sec, y_test_1_sec)

train_dataloader = DataLoader(dataset_train, batch_size=128, shuffle=True)
validation_dataloader = DataLoader(dataset_validation, batch_size=128, shuffle=True)
test_dataloader = DataLoader(dataset_test, batch_size=128, shuffle=True)

model = NeuralNetwork()

learning_rate = 0.001
epochs = 15

loss_fn = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
test_losses = []
train_losses = []

for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, loss_fn, optimizer)
    validation(validation_dataloader, model, loss_fn)
print("Test:")
test_loss = test(test_dataloader, model, loss_fn)
```

```python
epochs_arr = [i for i in range(1, epochs+1)]
plt.figure(figsize=(10, 6))

plt.plot(epochs_arr, test_losses, color='r', label='Validation Loss')
plt.plot(epochs_arr, train_losses, color='g', label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Learning Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_1_scaled = X_1_sec.copy()

X_1_scaled[columns] = scaler_1.transform(X_1_sec[columns])

predicted_y_1_sec = model(torch.tensor(np.array(X_1_scaled), dtype=torch.float32))

y_plot_predicted = predicted_y_1_sec.detach().numpy().squeeze(1)

y_plot_test = np.array(y_1_sec)

plt.figure(figsize=(25, 6))


plt.plot(y_plot_test, color='g', label='Testing value')
plt.plot(y_plot_predicted, color='r', label='Predicted Value')
plt.xlabel('Number')
plt.ylabel('MM256, %CH')
plt.title('Predicted and Testing Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error

print(f"MSE: {mean_squared_error(predicted_y_1_sec.detach().numpy().squeeze(1), np.array(y_1_sec))}")
print(f"MAPE: {mean_absolute_percentage_error(predicted_y_1_sec.detach().numpy().squeeze(1), np.array(y_1_sec))}")
print(f"MAE: {mean_absolute_error(predicted_y_1_sec.detach().numpy().squeeze(1),np.array(y_1_sec))}")

```
